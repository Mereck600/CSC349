# UI

This below here is a code I used ChatGPT to generate, using our logics. Many of the functions are redefined again here so that anyone can just run this one block of code without running anything before, to get it working. This is usually a bad practice because you ended up with multiple functions doing similar things and share the same name. For now, we do it for educational reasons. Run it, it should work.

````python
import numpy as np
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

ROW_COUNT = 6
COLUMN_COUNT = 7
EMPTY = 0
P1 = 1
P2 = 2

def create_board():
    return np.zeros((ROW_COUNT, COLUMN_COUNT), dtype=int)

def is_valid_location(board, col):
    # assumes row 0 is bottom? adjust if your code differs
    return board[ROW_COUNT-1, col] == EMPTY

def get_next_open_row(board, col):
    for r in range(ROW_COUNT):
        if board[r, col] == EMPTY:
            return r
    return None

def apply_move(board, col, piece):
    if not (0 <= col < COLUMN_COUNT) or not is_valid_location(board, col):
        return None
    r = get_next_open_row(board, col)
    board[r, col] = piece
    return r

def render_board_html(board):
    # Render with top row at top (so reverse rows for display)
    color = {EMPTY:"#ffffff", P1:"#e74c3c", P2:"#f1c40f"}
    border = "#2c3e50"
    cell = []
    cell.append("<table style='border-collapse:collapse;'>")
    for r in range(ROW_COUNT-1, -1, -1):
        cell.append("<tr>")
        for c in range(COLUMN_COUNT):
            cell.append(
                f"<td style='width:48px;height:48px;border:2px solid {border};"
                f"background:{border};text-align:center;'>"
                f"<div style='margin:auto;width:40px;height:40px;border-radius:50%;"
                f"background:{color[int(board[r,c])]};'></div></td>"
            )
        cell.append("</tr>")
    cell.append("</table>")
    return HTML("".join(cell))

# ---- UI wiring ----
board = create_board()
current_player = P1

out = widgets.Output()
status = widgets.HTML()

col_buttons = [widgets.Button(description=str(i), layout=widgets.Layout(width="48px"))
               for i in range(COLUMN_COUNT)]
btn_row = widgets.HBox(col_buttons)

def refresh():
    with out:
        clear_output(wait=True)
        display(render_board_html(board))
    status.value = f"<b>Turn:</b> {'Red (P1)' if current_player == P1 else 'Yellow (P2)'}"

def on_click_factory(col):
    def handler(_):
        global current_player
        r = apply_move(board, col, current_player)
        if r is None:
            status.value = "<b>Invalid move.</b> Column full."
            return
        # TODO: if you have winning_move(board, piece): check here
        current_player = P2 if current_player == P1 else P1
        refresh()
    return handler

for i, b in enumerate(col_buttons):
    b.on_click(on_click_factory(i))

reset_btn = widgets.Button(description="Reset", button_style="")
def reset(_):
    global board, current_player
    board = create_board()
    current_player = P1
    refresh()
reset_btn.on_click(reset)

display(widgets.VBox([widgets.HBox([widgets.HTML("<b>Click a column:</b>"), reset_btn]),
                      btn_row, status, out]))
refresh()
````
